# Algoritmo que faça a fatoração do número 1024. (Exemplo de fatoração: 6! = 6*5*4*3*2*1)

```python
# Função para calcular a fatoração de um número
def factorial(n):
    if n == 0 or n == 1:  # Fatoração de 0! ou 1! é 1
        return 1
    else:
        result = 1
        for i in range(2, n + 1):  # Multiplica todos os números de 2 até n
            result *= i
        return result

# Número a ser fatorado
number = 1024

# Calculando a fatoração
try:
    result = factorial(number)
    print(f"{number}! = {result}")
except OverflowError:
    print(f"A fatoração de {number} é muito grande para ser representada!")
```

# Fatoração em números primos

```python
def prime_factors(n):
    factors = []
    divisor = 2  # Começa com o menor número primo
    while n > 1:
        while n % divisor == 0:  # Enquanto o número for divisível pelo divisor
            factors.append(divisor)
            n //= divisor
        divisor += 1  # Passa para o próximo número
        if divisor * divisor > n:  # Otimização: não precisa testar acima da raiz quadrada de n
            if n > 1:
                factors.append(n)
            break
    return factors

# Número a ser fatorado
number = 1024

# Calculando os fatores primos
factors = prime_factors(number)
print(f"A decomposição em fatores primos de {number} é: {factors}")
```
